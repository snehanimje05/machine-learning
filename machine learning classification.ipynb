```python
import numpy as np
import pandas as  pd
import seaborn as snr
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/Breast_cancer_data.csv')
```

```python
data.head()
```

```python
x=data[['mean_radius','mean_texture','mean_perimeter','mean_area','mean_smoothness']]
y=data['diagnosis']
```
